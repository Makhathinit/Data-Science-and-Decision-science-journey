```python
df = pd.read_csv("/Users/thabisomakhathini/Downloads/Fuel.csv")
df 
```

```python
x_data = pd.DataFrame(df ,columns=['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','FUELCONSUMPTION_COMB_MPG'])
x_data
```

```python
y_data = pd.DataFrame(df ,columns=['CO2EMISSIONS'])
y_data
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size = 0.25)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
predict_y_test = regressor.predict(x_test)
predict_y_train = regressor.predict(x_train)
```

```python
from sklearn.metrics import r2_score
coefficient_of_dermination = r2_score(predict_y_train,y_train)
coefficient_of_dermination
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,predict_y_train, color ='m')
plt.title('CO2EMISSIONS vs FACTORS')
plt.xlabel('FACTORS')
plt.ylabel('CO2EMISSIONS')
```
